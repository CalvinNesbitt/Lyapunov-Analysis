```python
# Notebook for computing CLVs of the Lorenz 63 system
import numpy as np
import matplotlib.pyplot as plt 
from scipy.integrate import odeint
from scipy.linalg import expm
from mpl_toolkits.mplot3d import Axes3D
```

```python
# First we solve and plot L63
```

```python
# L63 Definition

#Parameters
a = 10.0
b = 8/3
c = 28.0

def L63(state, t, a=10.0, b=8/3, c=28.0):
    """ ODEs defining Lorenz 63 system.
    Parameter, state: current state.
    Parameter, t: time.
    Option, a: integer, parameter.
    Option, b: integer, parameter.
    Option, c: integer, parameter.
    """
    x, y, z = state
    dxdt = a * (y - x)
    dydt = (c * x) - y - (x * z)
    dzdt = (x * y) - (b * z)
    dudt = np.array([dxdt, dydt, dzdt])
    return dudt
```

```python
# Time Parameters

dt = 0.01 # Time step size
stepsA = 100 # Controls convergence to BLVS
stepsB = 30 # Number of points we want CLVs at
stepsC = 100 # Controls convergence of A- matrix, (coefficients of CLVs in BLV basis)
steps = stepsA + stepsB + stepsC # Number of steps
time = np.arange(0, steps * dt, dt)
```

```python
# Solving L63

IC = np.array([1.0, 1.0, -3.0]) 
solution = odeint(L63, IC, time)

# Plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(solution[:, 0], solution[:, 1], solution[:, 2])
plt.show()

print(solution.shape)
```

```python
# Jacobian Matrix

def Jacobian(location, parameters=[10.0, 10/3, 28.0]):
    """Returns the jacobian matrix at a given point of L63.
    Param, location, array: [x, y, z] coordinates.
    Option, parameters, L63 parameters.
    """
    [a, b, c] = parameters
    [x, y, z] = location
    row1 = np.array([-a, a, 0])
    row2 = np.array([c - z, -1, -x])
    row3 = np.array([y, x, -b])
    J = np.array([row1, row2, row3])
    return J   

# Jacobian at each point along trajectory

jacobians = np.zeros((3, 3, time.shape[0])) # Time series of Jacobian matrix. Indexed as (row, column, time)
for t in range(time.shape[0]):
    jacobians[:, :, t] = Jacobian(solution[t, :])

step = 100 # Print test
print(f'At step {step} (time {step * dt}), solution is\n\n {solution[step, :]}\n\nJacobian is\n\n{jacobians[:, :, step]}')
```

```python
# Propagator

k = 1 # Number of steps Propagator pushes dynamics forward
#CAN'T HAVE K NEQ 1 UNTIL YOU SORT OUT INDEXING AND STORAGE BELOW

def Propagator(step, k, jacobians):
    """ Propagator matrix F(step, step + k). Evolves solutions of TLE forward in time.
    Param, time1, integer: step we're caluclating propator at
    Param, k, how many steps we push forward.
    Option, parameters, L63 parameters.
    """
    F = jacobians[:, :, step]
    for i in range(1, k):
        F = np.matmul(F, jacobians[:, :, step + i])
    return F
```

```python
# Implementing the Ginelli Algorithm
```

```python
# Step 1: Getting the BLVs

# Bennetin Stepping

M0 = np.identity(3) # Take identity matrix as initial condition for matrix equation
oldM = M0 #Initialising oldQ
for i in range(0, stepsA, k):
    F = Propagator(i, k, jacobians) # Propagator
    M = np.matmul(F, oldM) # Pushing dynamics forward
    Q, R = np.linalg.qr(M) # Performing QR decomposition
    oldM = Q
    #if i in range(stepsA - 8, stepsA): # Print statement to check convergence
        #print(f'After {i} steps.\n')
        #print(f'Q is: \n\n {Q}\n\n')
        #print(f'R is: \n\n {R}\n') # In case you want to check R's also
```

```python
# Step 2: More Benettin steps
# Store both BLVs and R's.

Qs = np.zeros((3, 3, int(stepsB/k))) # Array to store BLV matrices. Indexed via (row, column, time)
Rs = np.zeros((3, 3, int(stepsB/k))) # Array to stor R matrices

oldM = Q #Initialising
for i in range(0, stepsB, k):
    F = Propagator(stepsA + i, k, jacobians) # Propagator
    M = np.matmul(F, oldM) # Pushing dynamics forward
    Q, R = np.linalg.qr(M) # Performing QR decomposition
    Qs[:,:, i], Rs[:, :, i] = Q, R
    oldM = Q
```

```python
# Step 3: More Benettin steps
# Now we stores R's only. 

Rs2 = np.zeros((3, 3, int(stepsC/k))) # Array to store R matrices

oldM = Qs[:,:, stepsB - 1] #Initialising oldM
for i in range(0, stepsC, k):
    F = Propagator(stepsA + stepsB + i, k, jacobians) # Propagator
    M = np.matmul(F, oldM) # Pushing dynamics forward
    Q, R = np.linalg.qr(M) # Performing QR decomposition
    Rs2[:, :, i] =  R
    oldM = Q
```

```python
# Step 4: Time to go back, converging to A- matrix
# Will also calculate LEs

# Initialise an upper triangular matrix
A = 3 * np.triu(M0)
A[0,1] =1

# We will evolve upper triangular matrix backward to timeB using the stored R's from Step 3
# This should converge to 'A-' of QR decomposition of CLVs "Gamma = phi^- A-"

LEs = np.zeros((3, int(stepsC/k)))

oldA = A
for i in range(0, stepsC, k):
    
    # Pushing A- backwards with R's
    Rinv = np.linalg.inv(Rs2[:, :, stepsC - i - 1])
    newA = np.matmul(Rinv, oldA)
    oldA = newA
    
    # Normalises A's to prevent overflow
    norms = np.linalg.norm(oldA, axis=0, ord=2) # L2 of column norms
    print(np.log(norms)/k)
    #norms = np.array([oldA[0, 0], oldA[1, 1], oldA[2, 2]]) # Keeping diaognal elements equal to 1
    oldA = oldA/norms 
    
    # Checking convergence
    #if i in range(timeC - 10, timeC):
        #print(f'A({timeC - i - 1})\n{oldA})\n\n')
        

Aminus = oldA


print(f'Lyapunov exponents are {np.log(norms)/k}\n\n') #L.E.'s
```

```python
# Step 5: Keep going back, finding CLVs

CLVs = np.zeros((3, 3, stepsB)) # Array to store CLVs. Indexed via (row, column, time)

oldA = Aminus

for i in range(0, stepsB, k):
    
    # Pushing A- backwards with R's
    Rinv = np.linalg.inv(Rs[:, :, stepsB - i - 1])
    newA = np.matmul(Rinv, oldA)
    
    # Normalise A-'s to prevent overflow
    norms = np.linalg.norm(newA, axis=0, ord=2) # L2 of column norms
    newA = newA/norms 
    
    # Calculate CLV, using A- and BLV
    BLV = Qs[:,:, stepsB - i - 1]
    CLVs[:, :, stepsB - i - 1] = np.matmul(BLV, newA)
    
    oldA = newA
    
    # Checking final steps
    if i in range(stepsB - 5, stepsB):
        print(f'CLVs({stepsB - i - 1})\n{CLVs[:, :, stepsB - i - 1]})\n\n')
```
